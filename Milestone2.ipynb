```python
import pandas as pd
import numpy as np

# Optional: display all columns
pd.set_option("display.max_columns", None)
```

```python
df = pd.read_csv(r"C:\CODING\PandasApp\azure_synthetic_dataset_5000.csv")

# for 20 data 
# df = pd.read_csv(r"C:\CODING\PandasApp\azure_synthetic_dataset_5000.csv").head(20)

# Display dataset shape and first 5 rows
print("Dataset Shape:", df.shape)
df.head()
```

```python
# Converting timestamp column from string to datetime format
df["timestamp"] = pd.to_datetime(df["timestamp"])

# Sort dataset by timestamp 
df = df.sort_values("timestamp").reset_index(drop=True)
print(df)
```

```python
#creating time features
df["hour"] = df["timestamp"].dt.hour
df["day"] = df["timestamp"].dt.day
df["weekday"] = df["timestamp"].dt.weekday

print(df)
```

```python
# df = df.sort_values(["region", "timestamp"])

df["lag_1_usage"] = df.groupby("region")["units_used"].shift(1)
df["lag_7_usage"] = df.groupby("region")["units_used"].shift(7)
df = df.reset_index(drop=True)
print(df.head(20))
```

```python
# Convert numeric columns to proper numeric format (if needed)
df["units_used"] = pd.to_numeric(df["units_used"], errors="coerce")
df["provision_capacity"] = pd.to_numeric(df["provision_capacity"], errors="coerce")
df["cost_use"] = pd.to_numeric(df["cost_use"], errors="coerce")
df["availability"] = pd.to_numeric(df["availability"], errors="coerce")

# Check for missing values
print(df.isnull().sum())
```

Demand-Driving Feature Engineering

Usage Trend (Rolling Mean)

```python

# Calculate rolling mean on correct column
df["rolling_mean_3"] = df["units_used"].rolling(window=3, min_periods=1).mean()

print(df)
```

# Capacity Utilization Ratio

```python
# Calculate how much of provisioned capacity is being used
df["capacity_utilization"] = df["units_used"] / df["provision_capacity"]

df.head()
```

# Cost Efficiency Metric

```python
# Calculating cost per unit
df["cost_per_unit"] = df["cost_use"] / df["units_used"]

df.head()
```

```python
# Availability Risk Indicator
# Convert availability to risk score

df["availability_risk"] = 1 - df["availability"]

df.head()
```

```python
# Seasonality & Time-Based Features
# Extract time-based features from timestamp
df["month"] = df["timestamp"].dt.month
df["day_of_week"] = df["timestamp"].dt.dayofweek

# Create weekend flag (1 = weekend, 0 = weekday)
df["is_weekend"] = df["day_of_week"].apply(lambda x: 1 if x >= 5 else 0)

df.head()
```

```python
# Holiday & Maintenance Flags
# Ensure holiday and maintenance flags are integers
df["external_holiday"] = df["external_holiday"].astype(int)
df["internal_maintenance"] = df["internal_maintenance"].astype(int)

df.head()
```

```python
# Lag Features (Critical for Time-Series Models)

# Create lag features to capture previous demand values
df["demand_lag_1"] = df["units_used"].shift(1)
df["demand_lag_7"] = df["units_used"].shift(7)

df.head(10)
```

```python



threshold = df["units_used"].mean() + df["units_used"].std()
df["usage_spike"] = np.where(df["units_used"] > threshold, 1, 0)
print(df)
```

```python
#  Encode Categorical Variables
# df = pd.get_dummies(df, columns=["region", "service_type"], drop_first=True)
```

```python
# Encode Categorical Variables

# Convert categorical variables to numeric
df_encoded = pd.get_dummies(df, columns=["region", "service_type"], drop_first=True)

# Drop timestamp (not needed for modeling)
df_encoded = df_encoded.drop(columns=["timestamp"])

# Drop rows with NaN (from lag features)
df_encoded = df_encoded.dropna()

print("Final model-ready dataset shape:", df_encoded.shape)
df_encoded.head()
```

```python
# Save Final Dataset

# Save the processed dataset for modeling phase
df_encoded.to_csv("azure_model_ready_dataset.csv", index=False)

print("Model-ready dataset saved successfully!")
```
